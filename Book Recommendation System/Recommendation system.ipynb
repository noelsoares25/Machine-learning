## Popularity based Recommender system

```python
# merging ratings and books columns
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
```

```python
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df, on= 'Book-Title')
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings'] >=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df['Image-URL-M'][0]
```

# Collaborative Filtering Based Recommender System

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
most_reading_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(most_reading_users)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns = 'User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#find cosine similarity....basiccally by finding euclidian distance for all the books
similarity_score=cosine_similarity(pt)
# we get a 2D array
```

```python
# silimality score of each book with every other
similarity_score
```

```python
# function for suggesting 5 books
def recommend(book_name):
#     fetch index of book
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key= lambda x:x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
#         print(pt.index[i[0]])
#         print(books[books['Book-Title'] == pt.index[i[0]]])
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M']))
        
        data.append(item)
        
    return data
```

```python
recommend('The Notebook')
```
